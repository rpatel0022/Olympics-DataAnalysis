```python
import pandas as pd

# Path to the downloaded dataset directory
dataset_path = '/root/.cache/kagglehub/datasets/heesoo37/120-years-of-olympic-history-athletes-and-results/versions/2'

# Load the athletes_event.csv file into a DataFrame named df
df = pd.read_csv(f'{dataset_path}/athlete_events.csv')

# Load the noc_regions.csv file into a DataFrame named region_df
region_df = pd.read_csv(f'{dataset_path}/noc_regions.csv')

# Display the first few rows of each DataFrame to confirm they loaded correctly
display(df.head())
display(region_df.head())
```

```python
df = df[df['Season'] == 'Summer']
```

```python
df.shape
```

```python
region_df.head()
```

```python
df = df.merge(region_df, on='NOC', how='left')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df['Medal'].value_counts()
```

```python
df = pd.concat([df,pd.get_dummies(df['Medal'])], axis=1)
```

```python
df.head()
```

```python
medal_tally = df.drop_duplicates(subset=['Team', 'NOC', 'Games', 'Year', 'City', 'Sport', 'Event', 'Medal'])
```

```python
medal_tally.head()
```

```python
medal_tally = medal_tally.groupby('NOC').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Gold', ascending=False).reset_index()
```
